```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
df = pd.read_csv('../../data/NIPS_1987-2015.csv')
df.head(3)
```

```python
# Must remove the string field
relevant_df = df.iloc[:, 1:]
relevant_df.head(3)
```

```python
Z = linkage(relevant_df, 'ward')
```

```python
plt.figure()
dendrogram(Z)
plt.show()
```
